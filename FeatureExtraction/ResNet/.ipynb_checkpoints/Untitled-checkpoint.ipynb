```python
from feature_extraction import get_feature_from_images,get_sess_model,get_feature_by_sess_model
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def imgHandle(img_data,alaph=1.4,beta=0):
    img_copy = img_data.copy().astype(np.float32)
    for i in range(img_copy.shape[0]):
        for j,res in enumerate(img_copy[i]<=250):
            if res:img_copy[i][j] = img_copy[i][j]*alaph+beta
        img_copy[i][img_copy[i]>255] = 255
    return img_copy.astype(np.int)
```

```python
img_path = "/Users/apple/PycharmProjects/UsePython2/data_augment/FaceRecognitionByIDImage/Img/10/12.jpg"
t_img_path = "/Users/apple/PycharmProjects/UsePython2/data_augment/FaceRecognitionByIDImage/Img/10/3.jpg"
img_data = cv2.imread(img_path)
t_img_data = cv2.imread(t_img_path)
```

```python
plt.imshow(img_data[:,:,::-1])
```

```python
o_img_data = img_data.copy()
o_img_data = cv2.cvtColor(o_img_data,cv2.COLOR_BGR2HSV)
o_img_data[:,:,2] = imgHandle(o_img_data[:,:,2],alaph=0.8,beta=5)
o_img_data = cv2.cvtColor(o_img_data,cv2.COLOR_HSV2BGR)
plt.imshow(o_img_data[:,:,::-1])
```
