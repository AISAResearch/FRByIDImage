```python
### Projecting 3D model onto the the image
## 根据 ref_U --> project
"""
ref_U 里面保存了3D的人脸模型,在三维空间中x,y,z的 coordinate
保存的是坐标!

这里是根据原来的图像的相机矩阵 和 已经有的三维人脸模型 映射到二维平面
"""
threedee = np.reshape(ref_U, (-1, 3), order='F').transpose() # (3,224*224)
temp_proj = proj_matrix * np.vstack((threedee, np.ones((1, threedee.shape[1])))) # (3,4) * (4,224*224) --> (3,224*224)
project = np.divide(temp_proj[0:2, :], np.tile(temp_proj[2, :], (2,1)))
```

```python
# display a image
def img_show(img_data):
    plt.imshow(img_data[:,:,::-1])
    plt.show()
    return
```

```python
def warpImg_(img, t_height, t_width, prj, idx):
    new_img = np.zeros((t_height*t_width, 3))
    new_img = new_img + 255
    ## In case we have some points
    if prj.size != 0:
        pixels = cv2.remap(img, np.squeeze( np.asarray( prj[0,:] ) ).astype('float32'),\
         np.squeeze( np.asarray( prj[1,:] ) ).astype('float32'),  cv2.INTER_CUBIC)
        pixels = pixels[:,0,:]
        new_img[idx,:] = pixels
    else:
        print '> Projected points empty'
    new_img = new_img.reshape(( t_height, t_width, 3), order='F')
    new_img[new_img > 255] = 255
    new_img[new_img < 0] = 0
    new_img = new_img.astype('uint8')
    return new_img
```
